```python
df = pd.read_csv("despression.csv")
```

```python
del df["Timestamp"]
```

```python
df = df.rename(columns={'Which year are you in?': 'year', 'How are you feeling right now?': 'feeling','On a scale of 1-100, how would you express this feeling?':'feeling_scale','Your location ?':'location','Relationship status':'relation','Are you happy with your financial state?':'finance','How much have you succeeded to cope up with the environment of your educational institution?':'place_copeup','Understanding with your family members?':'family_understanding','Are you feeling pressure in your study or work right now?':'pressure','Are you satisfied with your academic result?':'result','Are you happy with your living place?':'accomodation','Who supports you when you are not succeeding in your academic life?':'supporter','Have you used any social media within the last 6 hours?':'social_media','Do you have inferiority complex?':'inferioty','Are you satisfied with your meal today?':'meal','Are you feeling sick/health issues today?':'illness','Have you done any recreational activity (sports, gaming, hobby etc.) today?':'had_fun','How long did you sleep last night?(in hours)':'sleep'})
```

```python
del df["feeling_scale"]
```

```python
df['place_copeup'] =  'CopeUpLevel'+df['place_copeup'].astype(str)
```

```python
df['finance'] =  'Financially_happy ? '+df['finance'].astype(str)
```

```python
df['pressure'] =  'Feeling pressure?'+df['pressure'].astype(str)
```

```python
df['result'] =  'Happy With result ?'+df['result'].astype(str)
```

```python
df['accomodation'] =  'Happy With living place ?'+df['accomodation'].astype(str)
```

```python
df['supporter'] =  'Supported by : '+df['supporter'].astype(str)
```

```python
df['social_media'] =  'Used social media : '+df['social_media'].astype(str)
```

```python
df['Do you have inferiority complex? '] =  'Inferiority Complex : '+df['Do you have inferiority complex? '].astype(str)
```

```python
df['meal'] =  'Had good meal? : '+df['meal'].astype(str)
```

```python
df['illness'] =  'Feeling ill? : '+df['illness'].astype(str)
```

```python
df['had_fun'] =  'Had fun? : '+df['had_fun'].astype(str)
```

```python
df['sleep'] = df['sleep'].astype(int)



for i,row in df.iterrows():
    if row['sleep'] >=0 and row['sleep']<=4:
        df.at[i,'sleeping'] = 'zero to four'
    elif row['sleep'] >4 and row['sleep']<=6:
        df.at[i,'sleeping'] = 'four to six'
    elif row['sleep'] >6 and row['sleep']<=8:
        df.at[i,'sleeping'] = 'six to eight'
    else:
        df.at[i,'sleeping'] = 'over six'
       
```
